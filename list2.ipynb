19. Assign each row into one of three categories concerning the average self-esteem (below 13, between 13 and 14.5, and above 14.5). Using pivoting, find for each category and each group number of rows and the average weight loss.

```python
bins = [0, 13, 14.5, np.inf]
names = ['< 13', '13 - 14.5', '> 14.5']
df['Average Self-Esteem Category'] = pd.cut(df['Average Self-Esteem'], bins, labels=names)
# df.groupby(['Average Self-Esteem Category', 'group'])[['wl1', 'wl2', 'wl3']].mean()

# without NaNs
pd.pivot_table(df, values=['wl1', 'wl2', 'wl3'], index=['Average Self-Esteem Category', 'group'], aggfunc=np.mean)
```
